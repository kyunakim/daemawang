# 강북문화정보도서관

```python
import pandas as pd
gwangjin_df = pd.DataFrame(columns = ['강좌명','대상','기간'])
gwangjin_df
```

```python
from bs4 import BeautifulSoup
import requests


#main_content > div > ul > li > dl > dt > a


ind = 0 #행번호


for i in range(1,36): #range안에는 페이지수, 넉넉하게 넣기 #여기는 15페이지까지 가겠네  
    url = 'https://www.gwangjinlib.seoul.kr/gjinfo/menu/10111/program/30026/lectureList.do?currentPageNo='+ str(i) +'&lectureIdx=0'
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36"}
    gwangjin = requests.get(url, headers = headers, verify=False)
    bs = BeautifulSoup(gwangjin.text, 'html.parser')
    
    for j in range(1,11):
        try:
            title = bs.select_one('#contents > div.boardWrap > table > tbody > tr:nth-child('+str(j)+') > td.ta_l > a').text
            target = bs.select_one('#contents > div.boardWrap > table > tbody > tr:nth-child('+str(j)+') > td:nth-child(3)').text 
            period =bs.select_one('#contents > div.boardWrap > table > tbody > tr:nth-child('+str(j)+') > td.mobileHide2').text
            gwangjin_df.loc[ind] = [title, target, period]
            ind += 1
        except Exception as e:
            print(e)



gwangjin_df
```

```python
gwangjin_df['기간'] = gwangjin_df['기간'].str.replace('\t', '')
gwangjin_df['기간'] = gwangjin_df['기간'].str.replace('\n', '')
gwangjin_df['기간'] = gwangjin_df['기간'].str.replace('\r', '')



gwangjin_df
```

```python
gwangjin_df['대상'] = gwangjin_df['대상'].str.replace('일반', '전연령')
gwangjin_df['대상'] = gwangjin_df['대상'].str.replace( '전체', '전연령') 
gwangjin_df['대상'] = gwangjin_df['대상'].str.replace('유아', '어린이')
gwangjin_df['대상'] = gwangjin_df['대상'].str.replace('초등', '어린이')
gwangjin_df['대상'] = gwangjin_df['대상'].str.replace('중등', '청소년')
gwangjin_df['대상'] = gwangjin_df['대상'].str.replace( '고등', '청소년')
```
