```python
import pandas as pd
df_h = pd.read_csv('gangbuk_h.csv',
                encoding="utf-8-sig")
df_o = pd.read_csv('gangbuk_o.csv',
                encoding="utf-8-sig")
df = pd.concat([df_h, df_o] , axis=0)
df.reset_index(inplace = True)
df.drop(columns ='index', axis = 1 , inplace = True)
```

```python
import re
target = df['행사명'].values #Series.values = array
title_list = target.tolist() # array > list
text = ' '.join(title_list) #list들을 공백으로 구분해서 연결
text = re.sub('[^가-힣]+',' ',text) #한글 이외는 공백으로 대체
```

```python
pd.set_option('display.max_row', 500)
```

```python
#konlpy(콘엘피) = 형태소 분석
from konlpy.tag import Okt
from collections import Counter
okt = Okt() #객체형성

morph = okt.pos(text) #형태소로 변환 #word= 단어 / tag=품사
morph

noun_list =['메타버스','북스타트','초등','성인','플렌테리어','메타버스','북스타트','초등','성인','플렌테리어','메타버스','북스타트','초등','성인','플렌테리어','메타버스','북스타트','초등','성인','플렌테리어','메타버스','북스타트','초등','성인','플렌테리어','메타버스','북스타트','초등','성인','플렌테리어','메타버스','북스타트','초등','성인','플렌테리어','메타버스','북스타트','초등','성인','플렌테리어']
stop_word = ['만들기','장비','그림책','메이킹','체험','인문학','동화','사이클링','역사','청소년','강연','마음','민주','혁명','어린이','환경','미디어','학교생활','인공','지능','그린','코딩','여행','지구','이야기','스마트','만남','태양광','코칭','소망','큐브','독립','수집가','진로','육아','세계','글쓰기','민주주의','자격증','전문가','에너지','제로','시민','힐링','일러스트레이터','버스킹','아트','비트','뮤직','악기','팝아트','소품','가족','문학','공감','이해','녹음','친환경','견학','플라스틱','공기','정화','식물','어르신','학부모','신재생에너지','인생','가구','모빌리티','지속','탐험','복폴딩','영상','웹툰','에코']
for word, tag in morph: #word= 단어 / tag=품사
    if(tag=='Noun') & (len(word)>1): #noun이고 단어가 1글자 이상인것(1글자는 보통 조사이기때문)
        if word in stop_word:
            noun_list.append(word)

count = Counter(noun_list) #단어별 숫자세주기


data =dict(count.most_common()) #빈도수 정리 내림차
series = pd.Series(data)
df2=pd.DataFrame(series, columns=['cnt'])

df2
```
