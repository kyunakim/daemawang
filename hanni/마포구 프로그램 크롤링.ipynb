```python
from bs4 import BeautifulSoup
import requests


#main_content > div > ul > li > dl > dt > a


ind = 0 #행번호


for i in range(1,14): #range안에는 페이지수, 넉넉하게 넣기 #여기는 15페이지까지 가겠네  
    url = 'https://mplib.mapo.go.kr/mcl/MENU1069/PGM3021/lectureList.do?currentPageNo='+str(i)+'&lectureId='
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36"}
    mapo = requests.get(url, headers = headers)
    bs = BeautifulSoup(mapo.text, 'html.parser')
    
    for j in range(1,11):
        try:
            no = bs.select_one('#contents > div.boardWrap.mt20 > table > tbody > tr:nth-child('+str(j) +') > td:nth-child(1)').text
            library = bs.select_one('#contents > div.boardWrap.mt20 > table > tbody > tr:nth-child('+str(j) +') > td:nth-child(2)').text
            title = bs.select_one('#contents > div.boardWrap.mt20 > table > tbody > tr:nth-child('+str(j) +') > td.title > a').text
            target = bs.select_one('#contents > div.boardWrap.mt20 > table > tbody > tr:nth-child('+str(j) +') > td:nth-child(5)').text
            period =bs.select_one('#contents > div.boardWrap.mt20 > table > tbody > tr:nth-child('+str(j) +') > td:nth-child(6)').text
            pay = bs.select_one('#contents > div.boardWrap.mt20 > table > tbody > tr:nth-child('+str(j) +') > td:nth-child(8)').text
            mapo_df.loc[ind] = [no, library, title, target, period, pay]
            ind += 1
        except:
            pass


mapo_df
```

```python
import re
target = df['title'].values #Series.values = array
title_list = target.tolist() # array > list
text = ' '.join(title_list) #list들을 공백으로 구분해서 연결
text = re.sub('[^가-힣]+',' ',text) #한글 이외는 공백으로 대체
```

```python
#konlpy(콘엘피) = 형태소 분석
from konlpy.tag import Okt
from collections import Counter
okt = Okt() #객체형성

morph = okt.pos(text) #형태소로 변환 #word= 단어 / tag=품사
morph

noun_list =['리터러시','상호이해','리터러시','리터러시','리터러시','리터러시','리터러시','상호이해','상호이해','상호이해','상호이해','상호이해','상호이해','메타 버스','메타 버스','메타 버스','메타 버스','메타 버스','메타 버스','메타 버스','리딩튜터','리딩튜터','리딩튜터']
stop_word = []
for word, tag in morph: #word= 단어 / tag=품사
    if(tag=='Noun') & (len(word)>1): #noun이고 단어가 1글자 이상인것(1글자는 보통 조사이기때문)
        if word not in stop_word:
            noun_list.append(word)

count = Counter(noun_list) #단어별 숫자세주기


data = dict(count.most_common()) #빈도수 정리 내림차
series = pd.Series(data)
df = pd.DataFrame(series, columns=['cnt'])

df[80:] #인덱스 확인해보면서 스탑워드에 추가시키기..
```
