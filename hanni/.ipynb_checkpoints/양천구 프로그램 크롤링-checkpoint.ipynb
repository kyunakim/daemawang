```python
from bs4 import BeautifulSoup
import requests


#main_content > div > ul > li > dl > dt > a


ind = 0 #행번호

for i in range(1,4): #range안에는 페이지수, 넉넉하게 넣기 #여기는 15페이지까지 가겠네  
    url = 'https://lib.yangcheon.or.kr/yclib/contents.do?a_num=32295531'
    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36"}
    yc = requests.get(url, headers = headers)
    bs = BeautifulSoup(yc.text, 'html.parser')
    
    for j in range(1,20):
        try:
            lib = bs.select_one('#board > div.table_blist > table > tbody > tr:nth-child('+str(j)+') > td:nth-child(3)').text
            title = bs.select_one('#board > div.table_blist > table > tbody > tr:nth-child('+str(j)+') > td.title > a').text
            yc_df.loc[ind] = [lib, title]
            ind += 1
        except:
            pass


yc_df
```

```python
import pandas as pd
yc_shin = pd.DataFrame(columns = ['lib','title'])
yc_shin
```

```python
from bs4 import BeautifulSoup
import requests


#main_content > div > ul > li > dl > dt > a


ind = 0 #행번호

url = 'https://lib.yangcheon.or.kr/libsin/contents.do?a_num=39957162'
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36"}
yc = requests.get(url, headers = headers)
bs = BeautifulSoup(yc.text, 'html.parser')
    
for j in range(1,11):
    try:
        lib = bs.select_one('#board > div.table_blist > table > tbody > tr:nth-child('+str(j)+') > td:nth-child(3)').text
            #board > div.table_blist > table > tbody > tr > td.title > a
            #board > div.table_blist > table > tbody > tr > td:nth-child(3)
            #board > div.table_blist > table > tbody > tr:nth-child(1) > td.title > a
        title = bs.select_one('#board > div.table_blist > table > tbody > tr:nth-child('+str(j)+') > td.title > a').text
        yc_shin.loc[ind] = [lib, title]
        ind += 1
    except:
        pass
        

yc_shin
```
