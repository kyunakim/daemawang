```python
import pandas as pd
sd_df = pd.DataFrame(columns = ['lib','title','target'])
sd_df
```

```python
from bs4 import BeautifulSoup
import requests


url = 'https://www.sdlib.or.kr/SD/edusat/list.do'
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36"}
sd = requests.get(url, headers = headers)
bs = BeautifulSoup(sd.text, 'html.parser')


ind=0
for j in range(1,21):
    try:
        lib = bs.select_one('#board > div.lesson.no_top > ul > li:nth-child('+str(j)+') > div.cont > p.cate').text
        title = bs.select_one('#board > div.lesson.no_top > ul > li:nth-child('+str(j)+') > div.cont > p.tit > a').text
        target = bs.select_one('#board > div.lesson.no_top > ul > li:nth-child('+str(j)+') > div.cont > div > dl:nth-child(3) > dd').text
        sd_df.loc[ind] = [lib, title, target]
        ind += 1
    except:
        pass


sd_df
```

```python
sd_list = [sd_df, kh_df, yd_df, rb_df, ss_df, cg_df, fore_df]
sd_all = pd.concat(sd_list, ignore_index=True)
sd_all
```

```python
sd = sd_all.copy()
```

```python
sd['lib'] = sd['lib'].str.replace('\t', '')
sd['lib'] = sd['lib'].str.replace('\n', '')
sd['lib'] = sd['lib'].str.replace('\r', '')
sd
```

```python
sd['title'] = sd['title'].str.replace('\t', '')
sd['title'] = sd['title'].str.replace('\n', '')
sd['title'] = sd['title'].str.replace('\r', '')
sd
```
