- 기사 제목과 링크를 데이터프레임으로 가져온 후 본문 내용에서 한글 단어만 출력하여 형태소 분석
- 최근 국내 'AI'와 관련하여 가장 많이 언급되는 단어에 대한 빈도분석 및 시각화
- AI 분야에서의 유의할 만한 트랜드 정보 및 인사이트 도출

```python
import time
import requests
import json
import pandas as pd
import re
import warnings
warnings.filterwarnings('ignore')
```

```python
import urllib.request as rq
from urllib.parse import quote_plus
from bs4 import BeautifulSoup

query = quote_plus('ai')
articles_df = pd.DataFrame(columns=['url', 'text'])
total_urls = []
total_texts = []

# 기사 URL 가져오기
for i in range(10):
    url = f'https://search.naver.com/search.naver?where=news&query={query}&start={10*i+1}'
    response = rq.urlopen(url)
    soup = BeautifulSoup(response, 'lxml')
    mask = soup.select('.list_news > li > div > div > div > div > a.info')
    for x in mask:
        if x.get_text() == '네이버뉴스':
            total_urls.append(x['href'])
# print(total_urls)
# 기사 정보 가져오기
for url in total_urls:
    response = rq.urlopen(url)
    soup = BeautifulSoup(response, 'lxml')

# 기사 본문 가져오기
    text = soup.select_one('#dic_area')
    if text == None:
        total_texts.append([])
    else:
        total_texts.append(text.text)

# articles_df 데이터프레임에 url 저장
articles_df['url'] = total_urls
# articles_df 데이터프레임에 본문 저장
articles_df['text'] = total_texts

articles_df
```

```python
from konlpy.tag import Okt
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
stopword = ['분야', '통해', '이번', '관련', '보고서', '지난', '기존', '적용',
            '과정', '예정', '인공', '지능', '활용', '기반', '진행', '시간', '위해']

okt = Okt()
noun = []
for i in articles_df.text:
    morphs = okt.pos(str(i))
    for j, k in morphs:
        if (len(j) > 1) & (k == 'Noun'):
            noun.append(j)
wordsbody = dict(Counter(noun).most_common())
for i in stopword:
    if i in wordsbody:
        wordsbody.pop(i)

print('기사 본문 WordCloud')
wc = WordCloud(font_path='malgun.ttf',
               background_color="white", max_font_size=60)
cloud = wc.generate_from_frequencies(wordsbody)
plt.figure(figsize=(12, 10))
plt.axis('off')
plt.imshow(cloud)
plt.show()
```
